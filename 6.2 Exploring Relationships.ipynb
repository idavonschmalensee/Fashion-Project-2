### Categorical Plots

```python
df.loc[(df['Price'] >= 30) & (df['Price'] < 60), 'Price category'] = 'Middle price'
```

```python
df['Price category'].value_counts(dropna = False)
```

```python
sns.set(style="ticks")
g = sns.catplot(x="Price", y="Age", hue="Price category", data=df)
```
